# Transfer Functions

## Linear

```

def linear(x):
  return x 
```

```
x=np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, linear(x))
```

## ReLu

```
# ReLu

def reLu(x):
  return np.array([xi if xi >= 0 else 0 for xi in x])
```

```
x=np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, reLu(x))
```

## Sigmoid

```
def sigmoid(x, lam=1):
  return 1/(1+np.exp(-lam*x))
```

```
x=np.arange(-10,10,0.005)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, sigmoid(x))
```

```
# With a different lambda
x=np.arange(-10,10,0.005)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, sigmoid(x,10))
```

## Exponential Linear Unit (ELU)

```
def eLU(x, alpha=1):
  return np.array([xi if xi>0 else alpha*(np.exp(xi)-1) for xi in x ])
```

```
x=np.arange(-10,10,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, eLU(x))
```

```
# With a different alpha
x=np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, eLU(x,5))
```

## LogSigmoid

```
def logsigmoid(x, lam=1):
  return np.log(1/(1+np.exp(-lam*x)))
```

```

x=np.arange(-10,10,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, logsigmoid(x))
```

```
# With a different lambda
x=np.arange(-10,10,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, logsigmoid(x,5))
```

## Tanh

```
def tanh(x, lam=1):
  return (np.exp(lam*x)-np.exp(-lam*x))/(np.exp(lam*x)+np.exp(-lam*x))
```

```
# With a different lambda
x=np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, tanh(x))
```

```
# With a different lambda
x=np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, tanh(x, 5))
```
